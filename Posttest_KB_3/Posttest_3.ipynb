# Import Library

```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("Application_Dataset.csv")
```

## Pie Chart

```python
kelamin = data.groupby('Applicant_Gender')[['Applicant_Gender','Status']].sum().sort_values(by=['Status'],ascending=False)
kelamin.reset_index(inplace = True)
```

```python
warna = ("#FF0000","#0000FF")
plt.figure(figsize=(10,10))
plt.pie(kelamin["Status"],labels=kelamin["Status"],startangle=90,autopct='%.1f%%', shadow=True, explode=(0.05,0), colors=warna)
plt.title("Pengajuan Kartu Kredit Berdasarkan Jenis Kelamin")
plt.show()
```

Dari grafik diatas bisa dilihat bahwa lebih banyak Perempuan yang mengajukan kartu kredit daripada Laki-laki

## Grafik Batang

```python
pekerjaan = data.groupby('Income_Type')[['Income_Type','Status']].sum().sort_values(by=['Status'],ascending=False).head(10)
pekerjaan.reset_index(inplace = True)
```

```python
plt.figure(figsize=(15,5))
plt.bar(pekerjaan['Income_Type'],pekerjaan['Status'],color= "#00FFFF")
plt.title("Pengajuan Kartu Kredit Berdasarkan Golongan Pekerja")
plt.xlabel('Jenis Pekerjaan')
plt.ylabel('Pengaju')
plt.show()
```

Dari grafik diatas bisa dilihat bahwa jenis pekerja yang paling banyak menjadi pengaju adalah Working dan yang paling sedikit adalah State Servant

## Heatmap

```python
sns.scatterplot(data = data, x = "Status", y = "Total_Bad_Debt")
plt.show()
```

Dari Scatter Plot diatas bisa dilihat bahwa pemilik utang buruk yang berjumlah diatas 30 utang tidak diterima saat mengajukan kartu kredit
